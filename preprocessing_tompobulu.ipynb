```python
# Agregasi berdasarkan tanggal (ds)
df_clean = (
    df
    .groupby('ds', as_index=False)
    .mean(numeric_only=True)
)

# Cek hasil
df_clean.head()
```

```python
len(df_clean), df_clean['ds'].nunique()
```

```python
df_clean = df_clean.sort_values('ds').reset_index(drop=True)
df_clean['ds'].diff().value_counts().head()
```

```python
df_clean['ds'] = pd.to_datetime(df_clean['ds'])
df_clean.dtypes
```

```python
df_clean.columns = (
    df_clean.columns
    .str.lower()
    .str.replace(' ', '_')
    .str.replace('(', '', regex=False)
    .str.replace(')', '', regex=False)
    .str.replace('/', '_', regex=False)
)
```

```python
df_clean['rain_sum_mm'].describe()
```

```python
# Urutkan kolom
cols = ['ds', 'rain_sum_mm'] + [
    col for col in df_clean.columns if col not in ['ds', 'rain_sum_mm']
]

df_clean = df_clean[cols]
df_clean.head()
```

```python
df_clean.info()
```

```python
df_fe = df_clean.copy()

df_fe['year'] = df_fe['ds'].dt.year
df_fe['month'] = df_fe['ds'].dt.month
df_fe['day'] = df_fe['ds'].dt.day
df_fe['dayofyear'] = df_fe['ds'].dt.dayofyear

df_fe.head()
```

```python
df_fe.info()
```

```python
cols = (
    ['ds', 'year', 'month', 'day', 'dayofyear', 'rain_sum_mm'] +
    [col for col in df_fe.columns if col not in 
     ['ds', 'year', 'month', 'day', 'dayofyear', 'rain_sum_mm']]
)

df_fe = df_fe[cols]
df_fe.head()
```

```python
df_fe[['year', 'month', 'day', 'dayofyear']].describe()
```

```python
# Target (variabel utama)
target = 'rain_sum_mm'

# Fitur awal (numerik, kecuali ds)
initial_features = df_fe.select_dtypes(include=['int64', 'float64']).columns.tolist()
initial_features.remove(target)

initial_features
```

```python
# Hitung korelasi Pearson dengan target
corr = df_fe[initial_features + [target]].corr()[target].sort_values(ascending=False)

corr
```

```python
pearson_selected = corr[abs(corr) >= 0.1].index.tolist()
pearson_selected.remove(target)

pearson_selected
```

```python
from sklearn.feature_selection import VarianceThreshold

# Data fitur (tanpa ds dan target)
X = df_fe[initial_features]

# Variance Threshold
selector = VarianceThreshold(threshold=0.01)
selector.fit(X)

# Fitur yang lolos
variance_selected = X.columns[selector.get_support()].tolist()

variance_selected
```

```python
removed_features = set(initial_features) - set(variance_selected)
removed_features
```

```python
final_features = list(set(pearson_selected) & set(variance_selected))
final_features
```

```python
df_final = df_fe[['ds', target] + final_features]
df_final.head()
```
